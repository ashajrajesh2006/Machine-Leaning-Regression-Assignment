```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first='True')
```

```python
dataset
```

```python
dataset.columns
```

```python
independant= dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independant
```

```python
dependant=dataset[['charges']]
dependant
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independant, dependant, test_size=0.30, random_state=0) 
```

```python
from sklearn.tree import DecisionTreeRegressor # Procedure max_features='log2'
regressor=DecisionTreeRegressor(criterion='friedman_mse',splitter='random', max_features='sqrt')
regressor=regressor.fit(X_train,Y_train) # Model Creation
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
Y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test, Y_pred)
```
