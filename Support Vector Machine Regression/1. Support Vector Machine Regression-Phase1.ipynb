```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first='True')
```

```python
dataset 
```

```python
dataset.columns
```

```python
independant= dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independant
```

```python
dependant=dataset[['charges']]
dependant
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independant, dependant, test_size=0.30, random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVR # procedures
regressor=SVR(kernel="rbf",C=3000,epsilon=0.1) # procedures
regressor.fit(X_train,Y_train) # the output will create a model
```

```python
Y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test, Y_pred)
```

```python
r_score
```

```python
# Saving the model
import pickle 
filename="finalized_model_SVM.sav" 
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_SVM.sav",'rb'))
result=loaded_model.predict([[9945,125,12567,1,0]])
```
